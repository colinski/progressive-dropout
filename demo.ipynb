```python
import torch
from progressive_dropout import ProgressiveDropout
import matplotlib.pyplot as plt
```

```python
def plot_channels(x):
    _, axs = plt.subplots(1, 16, figsize=(16,16))
    axs = axs.flatten()
    for img, ax in zip(x[0], axs):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
```

```python
pdrop = ProgressiveDropout()
x = torch.randn(1, 16, 30, 30) #random feature maps
```

```python
plot_channels(x)
```

```python
plot_channels(pdrop(x))
```

```python
plot_channels(pdrop(x))
```

```python
plot_channels(pdrop(x))
```
